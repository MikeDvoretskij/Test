```python
import numpy as np
import cv2
import json
from IPython.display import display
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
depth_map = np.load("intern_test_3D_room-main/data/depth_map.npy")
h_corners = np.load("intern_test_3D_room-main/data/h_corners.npy")
v_corners = np.load("intern_test_3D_room-main/data/v_corners.npy")
with open("intern_test_3D_room-main/data/test_object_lu.json", "r", encoding="utf-8") as file:
    test_object_lu = json.load(file)["walls"]
image = cv2.open("intern_test_3D_room-main/data/partition.jpg")
ceiling_height = 2.5
```

```python
display(image)
```

```python
depth_map = depth_map.transpose()
depth_map.shape
```

```python
z = []
for i in v_corners.round():
    z.append(depth_map[int(i[0]), int(i[1])])
z = np.array(z)
insert_index = v_corners.shape[1]
v_corners = np.insert(v_corners, insert_index, z, axis=1)
v_corners
```

# Первый вариант

```python
upper_corners = v_corners[0::2]
lower_corners = v_corners[1::2]
```

```python
segment_boundaries = [np.vstack([upper_corners[i], lower_corners[i], lower_corners[(i + 1) % len(lower_corners)], upper_corners[(i + 1) % len(lower_corners)]]) for i in range(len(lower_corners))]
segment_boundaries
```

```python
def plot_wall(wall_corners, ceiling_height):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    wall = Poly3DCollection([wall_corners], facecolors='gray', edgecolors='black', alpha=0.7)
    ax.add_collection3d(wall)

    ax.auto_scale_xyz([0, 1024], [0, 1024], [0, ceiling_height])

    plt.axis('off')
    plt.show()
```

```python
for i, segment_corners in enumerate(segment_boundaries):
    if i != 5:
        plot_wall(segment_corners, ceiling_height)
    else:
        point1 = np.array([segment_corners[0, 0], segment_corners[0, 1], 0])
        point2 = np.array([segment_corners[3, 0], segment_corners[3, 1], 0])

        corners_wall6 = [
            np.array([segment_corners[0, 0], segment_corners[0, 1], 0]),
            np.array([segment_corners[1, 0], segment_corners[1, 1], ceiling_height]),
            np.array([segment_corners[2, 0], segment_corners[2, 1], ceiling_height]),
            np.array([segment_corners[3, 0], segment_corners[3, 1], 0]),
            point1,
            point2
        ]

        plot_wall(corners_wall6, ceiling_height)
```

```python
def plot_floor(segment_boundaries, ceiling_height):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    vertical_corners = np.vstack(segment_boundaries)

    floor_corners = np.unique(vertical_corners, axis=0)

    floor_corners = np.vstack([floor_corners, floor_corners[0]])

    floor = Poly3DCollection([floor_corners], facecolors='gray', edgecolors='black', alpha=0.7)
    ax.add_collection3d(floor)

    ax.auto_scale_xyz([0, 1024], [0, 1024], [0, ceiling_height])

    plt.axis('off')
    plt.show()

plot_floor(segment_boundaries, ceiling_height)
```
